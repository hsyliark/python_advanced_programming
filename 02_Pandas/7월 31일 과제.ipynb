### Q3

- key1의 값을 기준으로 data1의 값을 분류하여 합계를 구한 결과를 시리즈가 아닌 데이터프레임으로 구한다. 

```python
np.random.seed(0)
df = pd.DataFrame({
    'key1': ['A', 'A', 'B', 'B', 'A'],
    'key2': ['one', 'two', 'one', 'two', 'one'],
    'data1': [1, 2, 3, 4, 5],
    'data2': [10, 20, 30, 40, 50]
})
df
```

```python
ans = df.data1.groupby(df.key1).sum()
df1 = pd.DataFrame(ans.index, columns=["key1"])
df2 = pd.DataFrame(ans.values, columns=["sum"])
pd.merge(df1, df2, left_index=True, right_index=True)
```

### Q5

- 팁의 비율이 요일과 점심/저녁 여부, 인원수에 어떤 영향을 받는지 살펴본다.
- 어떤 요인이 가장 크게 작용하는지 판단할 수 있는 방법이 있는가? 

```python
tips = sns.load_dataset("tips")
tips['tip_pct'] = tips['tip'] / tips['total_bill']
tips.head(10)
```

```python
tips.groupby(["day", "time", "size"])[["tip_pct"]].describe()
```

- 인원수가 많아질수록 팁의 비율이 적어지는 경향이 있음.
- 토, 일요일에는 점심에 식당을 찾는 손님이 없고, 점심/저녁 여부에 따라 팁의 비율이 달라진다고 보기는 어려움.
- 대체적으로 주중에 찾는 손님이 팁을 많이 주는 경향이 있음.

```python
tips.groupby(["day"])[["tip_pct"]].describe()
```

```python
tips.groupby(["time"])[["tip_pct"]].describe()
```

```python
tips.groupby(["size"])[["tip_pct"]].describe()
```

- 각 요인에 따라 판단해봤을 때 인원수가 팁의 비율에 가장 많은 요인을 미치는 것으로 보여진다.

### Q6

- 타이타닉 승객 데이터를 이용하여 다음 분석을 실시하라.

```python
titanic = sns.load_dataset("titanic")
titanic.head(10)
```

- 1. qcut 명령으로 세 개의 나이 그룹을 만든다. 

```python
titanic["age_group"] = pd.qcut(titanic["age"], 3, labels=["young", "middle", "old"])
```

```python
titanic.tail(10)
```

- 2. 성별, 선실, 나이 그룹에 의한 생존율을 데이터프레임으로 계산한다. 행에는 성별 및 나이 그룹에 대한 다중 인덱스를 사용하고 열에는 선실 인덱스를 사용한다. 

```python
titanic.pivot_table("survived",["sex","age_group"],"pclass",aggfunc='sum')
```

- 3. 성별 및 선실에 의한 생존율을 피봇 데이터 형태로 만든다.

```python
titanic.pivot_table("survived","sex","pclass",aggfunc='sum')
```
