## 4. 데이터프레임 고급 인덱싱

- 데이터프레임에서 특정한 데이터만 골라내는 것을 인덱싱(indexing)이라고 한다. 앞 절에서는 라벨, 라벨 리스트, 인덱스데이터(정수) 슬라이스의 3가지 인덱싱 값을 사용하여 인덱싱을 하는 방법을 공부하였다. 그런데 Pandas는 numpy행렬과 같이 쉼표를 사용한 (행 인덱스, 열 인덱스) 형식의 2차원 인덱싱을 지원하기 위해 다음과 같은 특별한 인덱서(indexer) 속성도 제공한다.

- 출처 : https://datascienceschool.net/view-notebook/704731b41f794b8ea00768f5b0904512/

- loc : 라벨값 기반의 2차원 인덱싱
- iloc : 순서를 나타내는 정수 기반의 2차원 인덱싱
- at: 라벨값 기반의 2차원 인덱싱 (한개의 스칼라 값만 찾는다)
- iat : 순서를 나타내는 정수 기반의 2차원 인덱싱 (한개의 스칼라 값만 찾는다)

### 4.1 loc 인덱서
    - 인덱스데이터 <br>
    - 인덱스데이터 슬라이스 <br>
    - 인덱스데이터 리스트 <br>
    - 같은 행 인덱스를 가지는 불리언 시리즈 (행 인덱싱의 경우) <br>
    - 또는 위의 값들을 반환하는 함수 <br>

```python
import pandas as pd
import numpy as np
import pandas_datareader as pdr
```

```python
df = pd.DataFrame(np.arange(10, 22).reshape(3, 4),
                  index=["a", "b", "c"],
                  columns=["A", "B", "C", "D"])
df
```

- 인덱싱값을 하나만 받는 경우

```python
df.loc['a']
```

```python
df.loc['b':'c']  # df['b':'c']
```

```python
df.loc[["b", "c"]]
```

```python
df[["b", "c"]]  # KeyError
```

```python
df.A > 15
```

```python
df.loc[df.A > 15]
```

```python
def select_rows(df):
    return df.A > 15
select_rows(df)
```

```python
df.loc[select_rows(df)]
```

```python
df.loc["A"]  # KeyError
```

```python
df.loc[["A", "B"]]  # KeyError
```

```python
df2 = pd.DataFrame(np.arange(10, 26).reshape(4, 4), columns=["A", "B", "C", "D"])
df2
```

```python
df2.loc[1:2]
```

- 인덱싱값을 행과 열 모두 받는 경우

```python
df.loc["a", "A"]   # 행,열
```

```python
df.loc["b":, "A"]
```

```python
df.loc["a", :]
```

```python
df.loc[["a", "b"], ["B", "D"]]
```

```python
df.loc[df.A > 10, ["C", "D"]]
```

### 4.2 iloc 인덱서
    - 라벨이 아니라 순서를 나타내는 정수(integer) 인덱스만 받음

```python
df
```

```python
df.iloc[0, 1]
```

```python
df.iloc[:2, 2]
```

```python
df.iloc[0, -2:]
```

```python
df.iloc[2:3, 1:3]
```

```python
df.iloc[-1]
```

```python
df.iloc[-1] = df.iloc[-1] * 2
df
```

    - 10가지 이상의 방법으로 특정한 행과 열을 선택한다. 

### 4.3 at, iat 인덱서
    - 하나의 스칼라 값을 뽑을 때만 가능

```python
%timeit df.loc["a", "A"]
```

```python
%timeit df.at["a", "A"]
```

```python
%timeit df.iloc[0, 0]
```

```python
%timeit df.iat[0, 0]
```
